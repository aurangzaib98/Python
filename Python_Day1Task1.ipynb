```python
#DAY 1 Task 1 (1)
name = input("Enter your name: ")
age = input("Enter your age: ")

print("Hello", name + "!")
print("You are", age, "years old. Welcome to the community centre!")
```

```python
#Day 1 Task 1 (2)
length = float(input("Enter the kitchen length in meters: "))
width = float(input("Enter the kitchen width in meters: "))

area = length * width
cost = area * 12


print(f"To tile a kitchen of {length}m x {width}m, you will need {area} square meters of tiles.")

print(f"It will cost £{cost:.2f}.")
```

```python
#Day 1 Task 1 (3)
celsius = float(input("Enter today's temperature in Celsius:"))

fahrenheit = (celsius * 9/5) + 32

print(f"{celsius}°C is equal to {fahrenheit}°F.")
```

```python
#Day 1 Task 1 (4)
#Version A
miles = float(input("Please enter the distance in Miles:"))

km = round(miles*1.60934,3)

print(f"{miles} Miles is ~ {km} Kilometers.")
```

```python
#Day 1 Task 1 (4)
#Version B
km = float(input("Please enter the distance in Kilometers: "))

miles = round(km/1.60934,2)

print(f"{km} Kilomerters is ~ {miles:.2f} Miles")
```

```python
print("Please enter 4 digit number and I will reverse it.")
number = int(input("Enter your number:"))

ones = number%10
tens = number//10%10
hundreds = number//100%10
thousands = number//1000

print(f"{ones}{tens}{hundreds}{thousands}")
```
