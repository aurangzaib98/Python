# While Loop

```python
number = 0
while number < 5:
  number = number + 1
  if number == 3:
    continue
  print(number)
else:
    print('No longer < 5')
```

```python
counter = 0
while counter < 10:
  print ('Parking Available')
  counter = counter + 1
  if counter >= 10:
    print('Parking Full')
```

```python
# Write a program to find and print multiples of 7 between 0 to 50.
i = 0
while i<=50:
  if i%7==0:
    print(f"{i} is multiple of 7.")
  i += 1
```

```python
secret_number = 8
tries_left = 3

while tries_left > 0:
    guess = int(input("Guess the secret number (1-10): "))

    if guess == secret_number:
        print("🎉 Correct! You win!")
        break
    elif guess < secret_number:
        print("Too low.")
    else:
        print("Too high.")

    tries_left -= 1
    print(f"Tries left: {tries_left}")

if tries_left == 0:
    print("😢 Game over. The correct number was", secret_number)
```

```python
#Calculator using for loop
for i in range(3):
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    operator = input("Enter operator (+, -, *, /): ")

    if operator == "+":
        print("Result:", num1 + num2)

    elif operator == "-":
        print("Result:", num1 - num2)

    elif operator == "*":
        print("Result:", num1 * num2)

    elif operator == "/":
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Cannot divide by zero")

    else:
        print("Invalid operator")
        break
```
